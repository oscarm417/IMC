```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('backtest.txt',delimiter=';').fillna('')
prods = df['product'].unique()
data = dict() 
for p in prods:
    data[p] = df[df['product']==p]
for k,v in data.items():
    cols = [i if 'day' in i.lower() or 'time' in i.lower() else i+"_"+k for i in v.columns.tolist()]
    v.columns = cols
    data[k] = v
from functools import reduce
c = reduce(lambda a,b: a.merge(b, on = 'timestamp'),[v for k,v in data.items()])
pnls = [i for i in c.columns.tolist() if 'profit_and_loss' in i ]
c[pnls].sum(axis = 1).plot()
```
