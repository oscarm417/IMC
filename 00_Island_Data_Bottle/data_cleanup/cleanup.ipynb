```python
import pandas as pd

trade_data = pd.read_csv('../COMBINED/all_trades_combined.csv', delimiter=';').fillna(0)
orderbook_data = pd.read_csv('../COMBINED/all_prices_combined.csv', delimiter=';').fillna(0)

trade_data_pearls = pd.DataFrame(columns=trade_data.columns)
trade_data_bananas = pd.DataFrame(columns=trade_data.columns)
trade_data_coconuts = pd.DataFrame(columns=trade_data.columns)
trade_data_pina_coladas = pd.DataFrame(columns=trade_data.columns)
trade_data_diving_gear = pd.DataFrame(columns=trade_data.columns)
trade_data_berries = pd.DataFrame(columns=trade_data.columns)

orderbook_data_pearls = pd.DataFrame(columns=orderbook_data.columns)
orderbook_data_bananas = pd.DataFrame(columns=orderbook_data.columns)
orderbook_data_coconuts = pd.DataFrame(columns=orderbook_data.columns)
orderbook_data_pina_coladas = pd.DataFrame(columns=orderbook_data.columns)
orderbook_data_diving_gear = pd.DataFrame(columns=orderbook_data.columns)
orderbook_data_berries = pd.DataFrame(columns=orderbook_data.columns)
orderbook_data_dolphin_sightings = pd.DataFrame(columns=orderbook_data.columns)




for idx, row in trade_data.iterrows():              # This is taking forever lmao, thank god you only have to run this once per day
    if row['symbol'] == 'PEARLS':
        trade_data_pearls.loc[idx] = row
    elif row['symbol'] == 'BANANAS':
        trade_data_bananas.loc[idx] = row
    elif row['symbol'] == 'COCONUTS':
        trade_data_coconuts.loc[idx] = row
    elif row['symbol'] == 'PINA_COLADAS':
        trade_data_pina_coladas.loc[idx] = row
    elif row['symbol'] == 'DIVING_GEAR':
        trade_data_diving_gear.loc[idx] = row
    elif row['symbol'] == 'BERRIES':
        trade_data_berries.loc[idx] = row

for idx, row in orderbook_data.iterrows():
    if row['product'] == 'PEARLS':
        orderbook_data_pearls.loc[idx] = row
    elif row['product'] == 'BANANAS':
        orderbook_data_bananas.loc[idx] = row
    elif row['product'] == 'COCONUTS':
        orderbook_data_coconuts.loc[idx] = row
    elif row['product'] == 'PINA_COLADAS':
        orderbook_data_pina_coladas.loc[idx] = row
    elif row['product'] == 'DIVING_GEAR':
        orderbook_data_diving_gear.loc[idx] = row
    elif row['product'] == 'BERRIES':
        orderbook_data_berries.loc[idx] = row
    elif row['product'] == 'DOLPHIN_SIGHTINGS':
        orderbook_data_dolphin_sightings.loc[idx] = row
        

trade_data_pearls = trade_data_pearls.reset_index()
trade_data_bananas = trade_data_bananas.reset_index()
trade_data_coconuts = trade_data_coconuts.reset_index()
trade_data_pina_coladas = trade_data_pina_coladas.reset_index()
trade_data_diving_gear = trade_data_diving_gear.reset_index()
trade_data_berries = trade_data_berries.reset_index()

orderbook_data_pearls = orderbook_data_pearls.reset_index()
orderbook_data_bananas = orderbook_data_bananas.reset_index()
orderbook_data_coconuts = orderbook_data_coconuts.reset_index()
orderbook_data_pina_coladas = orderbook_data_pina_coladas.reset_index()
orderbook_data_diving_gear = orderbook_data_diving_gear.reset_index()
orderbook_data_berries = orderbook_data_berries.reset_index()
orderbook_data_dolphin_sightings = orderbook_data_dolphin_sightings.reset_index()


trade_data_pearls = trade_data_pearls.drop(['symbol', 'index'], axis=1)
trade_data_bananas = trade_data_bananas.drop(['symbol', 'index'], axis=1)
trade_data_coconuts = trade_data_coconuts.drop(['symbol', 'index'], axis=1)
trade_data_pina_coladas = trade_data_pina_coladas.drop(['symbol', 'index'], axis=1)
trade_data_diving_gear = trade_data_diving_gear.drop(['symbol', 'index'], axis=1)
trade_data_berries =trade_data_berries.drop(['symbol', 'index'], axis=1)


orderbook_data_pearls = orderbook_data_pearls.drop(['product', 'index'], axis=1)
orderbook_data_bananas = orderbook_data_bananas.drop(['product', 'index'], axis=1)
orderbook_data_coconuts = orderbook_data_coconuts.drop(['product', 'index'], axis=1)
orderbook_data_pina_coladas = orderbook_data_pina_coladas.drop(['product', 'index'], axis=1)
orderbook_data_diving_gear = orderbook_data_diving_gear.drop(['product', 'index'], axis=1)
orderbook_data_berries = orderbook_data_berries.drop(['product', 'index'], axis=1)
orderbook_data_dolphin_sightings = orderbook_data_dolphin_sightings.drop(['product', 'index'], axis=1)





orderbook_data_pearls['timestamp'] = ((orderbook_data_pearls['timestamp'] / 100) + 1) + ((orderbook_data_pearls['day'] + 2) * 10000)
trade_data_pearls['timestamp'] = ((trade_data_pearls['timestamp'] / 100) + 1) + ((trade_data_pearls['day'] + 2) * 10000)

orderbook_data_bananas['timestamp'] = ((orderbook_data_bananas['timestamp'] / 100) + 1) + ((orderbook_data_bananas['day'] + 2) * 10000)
trade_data_bananas['timestamp'] = ((trade_data_bananas['timestamp'] / 100) + 1) + ((trade_data_bananas['day'] + 2) * 10000)

orderbook_data_coconuts['timestamp'] = ((orderbook_data_coconuts['timestamp'] / 100) + 1) + ((orderbook_data_coconuts['day'] + 2) * 10000)
trade_data_coconuts['timestamp'] = ((trade_data_coconuts['timestamp'] / 100) + 1) + ((trade_data_coconuts['day'] + 2) * 10000)

orderbook_data_pina_coladas['timestamp'] = ((orderbook_data_pina_coladas['timestamp'] / 100) + 1) + ((orderbook_data_pina_coladas['day'] + 2) * 10000)
trade_data_pina_coladas['timestamp'] = ((trade_data_pina_coladas['timestamp'] / 100) + 1) + ((trade_data_pina_coladas['day'] + 2) * 10000)

orderbook_data_diving_gear['timestamp'] = ((orderbook_data_diving_gear['timestamp'] / 100) + 1) + ((orderbook_data_diving_gear['day'] + 2) * 10000)
trade_data_diving_gear['timestamp'] = ((trade_data_diving_gear['timestamp'] / 100) + 1) + ((trade_data_diving_gear['day'] + 2) * 10000)

orderbook_data_berries['timestamp'] = ((orderbook_data_berries['timestamp'] / 100) + 1) + ((orderbook_data_berries['day'] + 2) * 10000)
trade_data_berries['timestamp'] = ((trade_data_berries['timestamp'] / 100) + 1) + ((trade_data_berries['day'] + 2) * 10000)

orderbook_data_dolphin_sightings['timestamp'] = ((orderbook_data_dolphin_sightings['timestamp'] / 100) + 1) + ((orderbook_data_dolphin_sightings['day'] + 2) * 10000)





trade_data_pearls = trade_data_pearls.groupby(['timestamp'], as_index=False).agg({'price': 'mean', 'quantity': 'sum'})
trade_data_bananas = trade_data_bananas.groupby(['timestamp'], as_index=False).agg({'price': 'mean', 'quantity': 'sum'})
trade_data_coconuts = trade_data_coconuts.groupby(['timestamp'], as_index=False).agg({'price': 'mean', 'quantity': 'sum'})
trade_data_pina_coladas = trade_data_pina_coladas.groupby(['timestamp'], as_index=False).agg({'price': 'mean', 'quantity': 'sum'})
trade_data_diving_gear = trade_data_diving_gear.groupby(['timestamp'], as_index=False).agg({'price': 'mean', 'quantity': 'sum'})#
trade_data_berries = trade_data_berries.groupby(['timestamp'], as_index=False).agg({'price': 'mean', 'quantity': 'sum'})




print("Done")
```

```python
#SAFE DATA AS CSV

trade_data_pearls.to_csv('pearls_trade_data_all.csv', sep=';', header=trade_data_pearls.columns, index=False)
trade_data_bananas.to_csv('bananas_trade_data_all.csv', sep=';', header=trade_data_bananas.columns, index=False)
trade_data_coconuts.to_csv('coconuts_trade_data_all.csv', sep=';', header=trade_data_coconuts.columns, index=False)
trade_data_pina_coladas.to_csv('pina_coladas_trade_data_all.csv', sep=';', header=trade_data_pina_coladas.columns, index=False)
trade_data_diving_gear.to_csv('diving_gear_trade_data_all.csv', sep=';', header=trade_data_diving_gear.columns, index=False)
trade_data_berries.to_csv('berries_trade_data_all.csv', sep=';', header=trade_data_berries.columns, index=False)

orderbook_data_pearls.to_csv('pearls_orderbook_data_all.csv', sep=';', header=orderbook_data_pearls.columns, index=False)
orderbook_data_bananas.to_csv('bananas_orderbook_data_all.csv', sep=';', header=orderbook_data_bananas.columns, index=False)
orderbook_data_coconuts.to_csv('coconuts_orderbook_data_all.csv', sep=';', header=orderbook_data_coconuts.columns, index=False)
orderbook_data_pina_coladas.to_csv('pina_coladas_orderbook_data_all.csv', sep=';', header=orderbook_data_pina_coladas.columns, index=False)
orderbook_data_diving_gear.to_csv('diving_gear_orderbook_data_all.csv', sep=';', header=orderbook_data_diving_gear.columns, index=False)
orderbook_data_berries.to_csv('berries_orderbook_data_all.csv', sep=';', header=orderbook_data_berries.columns, index=False)
orderbook_data_dolphin_sightings.to_csv('dolphin_sightings_orderbook_data_all.csv', sep=';', header=orderbook_data_dolphin_sightings.columns, index=False)
```
