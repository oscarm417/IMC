```python
#CLEANUP THE FILES

import pandas as pd
import numpy as np

trade_data = pd.read_csv('all_trades_combined.csv', delimiter=';').fillna(0)


trade_data_pearls = pd.DataFrame(columns=trade_data.columns)
trade_data_bananas = pd.DataFrame(columns=trade_data.columns)
trade_data_coconuts = pd.DataFrame(columns=trade_data.columns)
trade_data_pina_coladas = pd.DataFrame(columns=trade_data.columns)
trade_data_diving_gear = pd.DataFrame(columns=trade_data.columns)
trade_data_berries = pd.DataFrame(columns=trade_data.columns)
trade_data_ukuleles = pd.DataFrame(columns=trade_data.columns)
trade_data_baguettes = pd.DataFrame(columns=trade_data.columns)
trade_data_dip = pd.DataFrame(columns=trade_data.columns)
trade_data_basket = pd.DataFrame(columns=trade_data.columns)





for idx, row in trade_data.iterrows():              # This is taking forever lmao, thank god you only have to run this once per day
    if row['symbol'] == 'PEARLS':
        trade_data_pearls.loc[idx] = row
    elif row['symbol'] == 'BANANAS':
        trade_data_bananas.loc[idx] = row
    elif row['symbol'] == 'COCONUTS':
        trade_data_coconuts.loc[idx] = row
    elif row['symbol'] == 'PINA_COLADAS':
        trade_data_pina_coladas.loc[idx] = row
    elif row['symbol'] == 'DIVING_GEAR':
        trade_data_diving_gear.loc[idx] = row
    elif row['symbol'] == 'BERRIES':
        trade_data_berries.loc[idx] = row
    elif row['symbol'] == 'UKULELE':
        trade_data_ukuleles.loc[idx] = row
    elif row['symbol'] == 'BAGUETTE':
        trade_data_baguettes.loc[idx] = row
    elif row['symbol'] == 'DIP':
        trade_data_dip.loc[idx] = row
    elif row['symbol'] == 'PICNIC_BASKET':
        trade_data_basket.loc[idx] = row




trade_data_pearls = trade_data_pearls.reset_index()
trade_data_bananas = trade_data_bananas.reset_index()
trade_data_coconuts = trade_data_coconuts.reset_index()
trade_data_pina_coladas = trade_data_pina_coladas.reset_index()
trade_data_diving_gear = trade_data_diving_gear.reset_index()
trade_data_berries = trade_data_berries.reset_index()
trade_data_ukuleles = trade_data_ukuleles.reset_index()
trade_data_baguettes = trade_data_baguettes.reset_index()
trade_data_dip = trade_data_dip.reset_index()
trade_data_basket = trade_data_basket.reset_index()



trade_data_pearls = trade_data_pearls.drop(['symbol', 'index'], axis=1)
trade_data_bananas = trade_data_bananas.drop(['symbol', 'index'], axis=1)
trade_data_coconuts = trade_data_coconuts.drop(['symbol', 'index'], axis=1)
trade_data_pina_coladas = trade_data_pina_coladas.drop(['symbol', 'index'], axis=1)
trade_data_diving_gear = trade_data_diving_gear.drop(['symbol', 'index'], axis=1)
trade_data_berries =trade_data_berries.drop(['symbol', 'index'], axis=1)
trade_data_ukuleles =trade_data_ukuleles.drop(['symbol', 'index'], axis=1)
trade_data_baguettes =trade_data_baguettes.drop(['symbol', 'index'], axis=1)
trade_data_dip =trade_data_dip.drop(['symbol', 'index'], axis=1)
trade_data_basket =trade_data_basket.drop(['symbol', 'index'], axis=1)







trade_data_pearls['timestamp'] = ((trade_data_pearls['timestamp'] / 100) + 1) + ((trade_data_pearls['day'] + 2) * 10000)

trade_data_bananas['timestamp'] = ((trade_data_bananas['timestamp'] / 100) + 1) + ((trade_data_bananas['day'] + 2) * 10000)

trade_data_coconuts['timestamp'] = ((trade_data_coconuts['timestamp'] / 100) + 1) + ((trade_data_coconuts['day'] + 2) * 10000)

trade_data_pina_coladas['timestamp'] = ((trade_data_pina_coladas['timestamp'] / 100) + 1) + ((trade_data_pina_coladas['day'] + 2) * 10000)

trade_data_diving_gear['timestamp'] = ((trade_data_diving_gear['timestamp'] / 100) + 1) + ((trade_data_diving_gear['day'] + 2) * 10000)

trade_data_berries['timestamp'] = ((trade_data_berries['timestamp'] / 100) + 1) + ((trade_data_berries['day'] + 2) * 10000)

trade_data_ukuleles['timestamp'] = ((trade_data_ukuleles['timestamp'] / 100) + 1) + ((trade_data_ukuleles['day'] + 2) * 10000)

trade_data_baguettes['timestamp'] = ((trade_data_baguettes['timestamp'] / 100) + 1) + ((trade_data_baguettes['day'] + 2) * 10000)

trade_data_dip['timestamp'] = ((trade_data_dip['timestamp'] / 100) + 1) + ((trade_data_dip['day'] + 2) * 10000)

trade_data_basket['timestamp'] = ((trade_data_basket['timestamp'] / 100) + 1) + ((trade_data_basket['day'] + 2) * 10000)



```

```python
#SAFE DATA AS CSV

trade_data_pearls.to_csv('pearls_trade_data_all.csv', sep=';', header=trade_data_pearls.columns, index=False)
trade_data_bananas.to_csv('bananas_trade_data_all.csv', sep=';', header=trade_data_bananas.columns, index=False)
trade_data_coconuts.to_csv('coconuts_trade_data_all.csv', sep=';', header=trade_data_coconuts.columns, index=False)
trade_data_pina_coladas.to_csv('pina_coladas_trade_data_all.csv', sep=';', header=trade_data_pina_coladas.columns, index=False)
trade_data_diving_gear.to_csv('diving_gear_trade_data_all.csv', sep=';', header=trade_data_diving_gear.columns, index=False)
trade_data_berries.to_csv('berries_trade_data_all.csv', sep=';', header=trade_data_berries.columns, index=False)
trade_data_ukuleles.to_csv('ukuleles_trade_data_all.csv', sep=';', header=trade_data_ukuleles.columns, index=False)
trade_data_baguettes.to_csv('baguette_trade_data_all.csv', sep=';', header=trade_data_baguettes.columns, index=False)
trade_data_dip.to_csv('dip_trade_data_all.csv', sep=';', header=trade_data_dip.columns, index=False)
trade_data_basket.to_csv('basket_trade_data_all.csv', sep=';', header=trade_data_basket.columns, index=False)



```

```python
#GET UNIQUE TRADERS PER MARKET
import pandas as pd
import numpy as np


pearls_trade_data_all = pd.read_csv('pearls_trade_data_all.csv', delimiter=';').fillna(0)
pearl_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

bananas_trade_data_all = pd.read_csv('bananas_trade_data_all.csv', delimiter=';').fillna(0)
banana_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

coconuts_trade_data_all = pd.read_csv('coconuts_trade_data_all.csv', delimiter=';').fillna(0)
coconut_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

pina_coladas_trade_data_all = pd.read_csv('pina_coladas_trade_data_all.csv', delimiter=';').fillna(0)
pina_colada_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

diving_gear_trade_data_all = pd.read_csv('diving_gear_trade_data_all.csv', delimiter=';').fillna(0)
diving_gear_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

berries_trade_data_all = pd.read_csv('berries_trade_data_all.csv', delimiter=';').fillna(0)
berrie_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

ukuleles_trade_data_all = pd.read_csv('ukuleles_trade_data_all.csv', delimiter=';').fillna(0)
ukulele_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

baguette_trade_data_all = pd.read_csv('baguette_trade_data_all.csv', delimiter=';').fillna(0)
baguette_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

dip_trade_data_all = pd.read_csv('dip_trade_data_all.csv', delimiter=';').fillna(0)
dip_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])

basket_trade_data_all = pd.read_csv('basket_trade_data_all.csv', delimiter=';').fillna(0)
basket_traders = pd.DataFrame(columns=['timestamp', 'name', 'buys', 'sells', 'price'])



for idx, row in pearls_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    pearl_traders = pd.concat([pearl_traders, data], axis=0)

for idx, row in bananas_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    banana_traders = pd.concat([banana_traders, data], axis=0)

for idx, row in coconuts_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    coconut_traders = pd.concat([coconut_traders, data], axis=0)

for idx, row in pina_coladas_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    pina_colada_traders = pd.concat([pina_colada_traders, data], axis=0)

for idx, row in diving_gear_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    diving_gear_traders = pd.concat([diving_gear_traders, data], axis=0)

for idx, row in berries_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    berrie_traders = pd.concat([berrie_traders, data], axis=0)

for idx, row in ukuleles_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    ukulele_traders = pd.concat([ukulele_traders, data], axis=0)

for idx, row in baguette_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    baguette_traders = pd.concat([baguette_traders, data], axis=0)

for idx, row in dip_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    dip_traders = pd.concat([dip_traders, data], axis=0)

for idx, row in basket_trade_data_all.iterrows():  
    
    data = pd.DataFrame(data={'timestamp': [row['timestamp'], row['timestamp']], 'name' : [row['buyer'], row['seller']], 'buys' : [row['quantity'], 0], 'sells' : [0, -row['quantity']], 'price': [row['price'], row['price']]}, columns=['timestamp', 'name', 'buys', 'sells', 'price'])
    basket_traders = pd.concat([basket_traders, data], axis=0)
    

pearl_traders = pearl_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
banana_traders = banana_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
pina_colada_traders = pina_colada_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
coconut_traders = coconut_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
diving_gear_traders = diving_gear_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
berrie_traders = berrie_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
ukulele_traders = ukulele_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
baguette_traders = baguette_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
dip_traders = dip_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
basket_traders = basket_traders.groupby(['timestamp', 'name'], as_index=False).agg({'buys': 'sum', 'sells': 'sum', 'price': 'mean'})
```
